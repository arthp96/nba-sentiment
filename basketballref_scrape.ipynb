```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.safari.options import Options
from bs4 import BeautifulSoup
import string
```

```python
alphabet = list(string.ascii_lowercase)
driver = webdriver.Safari()
all_players = []
all_players_current = []
for letter in alphabet:
    link = 'https://www.basketball-reference.com/players/' + letter + '/'
    driver.get(link)
    source = driver.page_source
    soup = BeautifulSoup(source, 'html.parser')
    names = [tr.th.a.text for tr in soup.table.tbody.find_all('tr') if tr.th.a]
    current_names = [tr.th.strong.a.text for tr in soup.table.tbody.contents if tr != '\n' and tr.th.strong]

    all_players.extend(names)
    all_players_current.extend(current_names)
    #print(current_names)

driver.quit()
```

```python
teams_link = 'https://www.basketball-reference.com/teams/'
driver = webdriver.Safari()
driver.get(teams_link)
source = driver.page_source
driver.quit()

soup = BeautifulSoup(source, 'html.parser')

teams = []
for item in [line for line in soup.table.tbody.contents if line != '\n']: #soup.table.tbody.contents:
    if item['class'] == ['full_table']:
        teams.append(item.a.text)
```
